```python
# Numpy Arrays
a = np.array([1, 2, 3]) # 1D  array
b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float) # 2D array
c = np.array([[(1.5, 2, 3), (4, 5, 6)], [(3, 2, 1), (4, 5, 6)]]) # 3D array
d = np.array([[[[1.5, 2, 3], [4, 5, 6]], [[3, 2, 1], [4, 5, 6]]]]) # 4D array -> stacking of 3d arrays

print("1D array: \n", a, "\nShape: ", a.shape, "\n---\n")
print("2D array: \n", b, "\nShape: ", b.shape, "\n---\n")
# Shape: (2, 3) means -> 2 rows, 3 columns

print("3D array: \n", c, "\nShape: ", c.shape, "\n---\n")
# Shape: (2, 2, 3) means -> 2 blocks (sub-arrays), 2 rows, 3 columns

print("4D array: \n", d, "\nShape: ", d.shape)
# Shape:  (1, 2, 2, 3) means -> 1 depth layer, 2 sub-blocks, 2 rows, 3 columns
```

```python
# Standard Data Types
# integers -> signed : np.int64, np.int32, np.int16, np.int8
#           unsigned : np.uint64, np.uint32, np.uint16, np.uint8

# floats -> float16 (half precision), float32 (single), float64 (double)
#           float128 / float96 (extended precision)

# booleans -> bool_ (True/False)

# complex numbers -> complex64 (float32 real + imaginary), 
#                   complex128 (float64 real + imaginary)
#                   complex256 (on some platforms) 

# strings -> object_ 
```

```python
import numpy as np

# Helper to print type info
def show(name, arr):
    print(f"{name:15} dtype={str(arr.dtype):10} shape={arr.shape}  example={arr}")

# 1) Signed integers
a_int8  = np.array([1, -2], dtype=np.int8)   # small-range integer storage
a_int16 = np.array([1000, -2000], dtype=np.int16)
a_int32 = np.array([10_000_000, -5], dtype=np.int32)
a_int64 = np.array([10**12, -3], dtype=np.int64)  # large integers
show("int8",  a_int8)
show("int16", a_int16)
show("int32", a_int32)
show("int64", a_int64)
print(np.issubdtype(a_int32.dtype, np.signedinteger))  # True
print("\n---")

# 2) Unsigned integers
u8  = np.array([0, 255], dtype=np.uint8)   # bytes, e.g., image channels
u16 = np.array([0, 65535], dtype=np.uint16)
show("uint8",  u8)
show("uint16", u16)
print(np.issubdtype(u8.dtype, np.unsignedinteger))  # True
print("\n---")

# 3) Pointer-sized index type
idx = np.array([0,1,2], dtype=np.intp)  # used by NumPy for indexing/lengths
show("intp", idx)
print("\n---")

# 4) Floating point
f16 = np.array([0.1, 1.5], dtype=np.float16)   # memory-saving, low precision
f32 = np.array([0.1, 1.5], dtype=np.float32)   # ML / GPU common
f64 = np.array([0.1, 1.5], dtype=np.float64)   # default double precision
show("float16", f16)
show("float32", f32)
show("float64", f64)
print(np.finfo(f32.dtype))   # precision/limits
print("\n---")
```

```python
# 5) Extended float (platform-dependent)
try:
    ld = np.array([1.0, 2.0], dtype=np.longdouble)
    show("longdouble", ld)
except Exception as e:
    print("longdouble not supported on this platform:", e)
print("\n---")

# 6) Complex types
c64  = np.array([1+2j, 3+4j], dtype=np.complex64)   # complex with float32 parts
c128 = np.array([1+2j, 3+4j], dtype=np.complex128)  # complex with float64 parts
show("complex64", c64)
show("complex128", c128)
print(np.iscomplexobj(c128))
print("\n---")

# 7) Boolean
boolean = np.array([True, False, True], dtype=np.bool_)
show("bool_", boolean)
print("\n---")

# 8) Byte-string and Unicode
s_bytes = np.array([b"abc", b"def"], dtype='S5')  # fixed-width bytes
s_unicode = np.array(["α", "β"], dtype='U4')      # fixed-width unicode
show("S5 (bytes)", s_bytes)
show("U4 (unicode)", s_unicode)
print("\n---")

# 9) Object (Python objects)
obj = np.array([{"a":1}, (1,2)], dtype=object)  # mixed types, Python objects
show("object_", obj)
print(isinstance(obj[0], dict))
print("\n---")

# 10) Datetime / Timedelta
dt = np.array(['2020-01-01', '2021-06-01'], dtype='datetime64[D]')
td = np.array([1, 30], dtype='timedelta64[D]')
show("datetime64[D]", dt)
show("timedelta64[D]", td)
print(dt.dtype, td.dtype)
print("\n---")

# 11) Structured / record dtype
rec = np.array([(1, 2.5, 'a'), (2, 3.5, 'b')],
               dtype=[('id','i4'), ('val','f4'), ('char','U1')])
print("record dtype:", rec.dtype)
print("fields access:", rec['id'], rec['val'], rec['char'])
print("\n---")

# 12) Void / raw bytes
v = np.array([b"\x00\x01\x02"], dtype='V3')  # raw 3 bytes
show("void V3", v)
print("\n---")

# 13) Aliases and checks
print("np.int_: ", np.dtype(int))
print("np.float_: ", np.dtype(float))
print("issubdtype examples:",
      np.issubdtype(np.dtype('int32'), np.integer),
      np.issubdtype(np.dtype('float64'), np.floating))
print("\n---")

# 14) Programmatic listing (dump names/aliases)
# Uncomment to print many available type names/aliases on your platform:
print(sorted(set(np.sctypeDict.keys())))
```

```python
arr_3d = np.array([[[3, 6, 2], [5, 2, 7], [2, 4, 1],],
                   [[2, 2, 5], [4, 2, 1], [7, 9, 2]]])
print(arr_3d.shape)

copy = arr_3d.copy()

def sort_across(num: int, arr: np.array) -> np.array:
    # axis=0 sorts along the first axis (across the two blocks) -> for each (row, col) position it sorts the 2 values from the two blocks.
    # axis=1 sorts along the second axis (rows) -> for each (block, col) position it sorts the 3 values across rows.
    # axis=2 sorts along the third axis (columns/elements in each row) -> for each (block, row) position it sorts the 3 values in that row.
    local_copy = arr.copy()
    if num>2:
        raise ValueError("Out of range for a 3d array.")
    local_copy.sort(axis=num)
    return local_copy
    

print("Original Array: ", arr_3d, "\n")
res = sort_across(int(input("Enter the integer: ")), arr_3d)
print("Sorted array: ", res)
```

```python
# Subsetting, Slicing, Indexing
print("Array a:", a)
print("Array c:\n", c, "\n")
print("\n---\n")


# 1. Basic Indexing -> select the element (or sub-array) at index i along the first axis
print(a[2])
print(c[1])
print("\n---\n")

# 2. Multi-axis integer indexing
# NOTE: c is a 3D array -> block, rows, columns
print(c[1, 1]) # select the element at block 1, row 1
print(c[0, 1, 0]) # select the element at block 0, row 1, column 0
print("\n---\n")
```

```python
# 3. Slicing -> a[start:stop:step] -> returns a view (not a copy) when possible
print(a[0:3:2])
print("\n---\n")

# 4. Multi-axis slicing -> slicing on each axis is applied independently; result may be a view.
print(c[0, 0:2, 1]) # select element(s) from block 0, rows 0-1, from column 1
print(c[0:1, 1, 0]) # select element(s) from block 0, row 1, column 0
print(c[1, 0, 1:2]) # select element(s) from block 1, row 0, column 1
print("\n---\n")

# 5. Negative step slicing (reversal) -> traverse the axis in reverse order; returns a view
print(a[::-1])
print(c[::-1]) # reverses blocks
```

```python
# 6. Ellipsis -> c['n', ...] -> select any index on the first axis, and all remaining axes
print(c[1,...]) # select first axis -> here represents block 1, rest same
print("\n---\n")

# 7. Boolean Indexing (Masking) -> create a boolean mask by evaluating a condition
# returns a copy containing only the selected elements (not a view). Shape is 1‑D regardless of original.
print(a[a<4])
print(c[c<5])
print(c[c!=2])
print("\n---\n")
```

```python
# 8. Fancy Indexing 
print("Array b: \n", b)
print("\n---\n")

# 8.1 paired index arrays -> take elements at coordinates formed by pairing corresponding indices
# returns a 1‑D array with one item per index-pair; integer indexing always returns a copy
print(b[[1, 0, 1, 0],[0, 1, 2, 0]]) # index pairs: (1,0), (0,1), (1,2), (0,0)
print(b[[1, 1, 1], [0, 2, 1]]) # (1,0), (1,2), (1,1)
print("\n---\n")

# 8.2 row/column selection separately
print(b[[1,0,1,0]][:, [0,1,2,0]])
# first b[[1,0,1,0]] selects those rows (order preserved)
# then, for each row, select columns 0,1,2,0 
print(b[[1, 0]][:, [2, 0]]) 
# rows 1 and 0 respectively then: row 1: (1,2), (1,0)
#                                 row 2: (0,2), (0,0) 
```

```python
# Array manipulation
print("Array b: \n", b)
print("Array c: \n", c)
print("\n---\n")

# 1. Transposing an array -> reordering of axes
print(np.transpose(b)) # interchange rows and columns in 2D arrays
print(c.T) # random axes permutation
print(c.transpose(1, 2, 0)) # new order = rows as blocks, columns as rows, blocks as columns
print("\n---\n")
```

```python
# 2. Changing array shape
print(b.ravel()) # returns a 1-D view (shares memory when possible)
print(b.flatten()) # returns a 1-D copy (always new memory)
print(b.reshape(2, 3)) # reshape to new dimensions (view if possible)
print(b.T) # transpose (swap axes)

# 3. Adding/removing elements
# 1-D arrays
print("1-D array: \n", a)
print(np.append(a, [4, 5, 6])) # append elements to the end of an array
print(np.insert(a, 1, 100)) # insert 100 at index 1
print(np.delete(a, 0)) # remove element at index 0
print(np.delete(a, [0, 2])) # remove multiple indices

# 2-D arrays -> for higher order matrices, use axis to control row vs column behaviour and so on
print("2-D array: \n", b)
print(np.append(b, [[5, 6, 3]], axis=0)) # add a new row
print(np.delete(b, 0, axis=1)) # delete first column
```

```python
# 4. Combining Arrays
# 1-D arrays
a1 = np.array([1, 2, 3])
b1 = np.array([2, 7, 9])
print(np.concatenate([a1, b1])) # join along existing axis
print(np.stack([a1, b1])) # join along a new axis
print(np.vstack([a1, b1])) # stack vertically (row-wise)
print(np.hstack([a1, b1])) # stack horixontally (column-wise)

# 2-D arrays
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])
print(np.concatenate([m1, m2]), axis=0) # stack rows
print(np.concatenate([m1, m2]), axis=1) # stack columns

# 5. Splitting Arrays
p = np.arange(9)
print(np.split(p, 3)) # split into 3 EQUAL parts

q = np.arange(10)
print(np.array_split(b, 3)) # split into 3 parts, allows UNEQUAL sizes

s = np.arange(16).reshape(4, 4)
print(np.hsplit(s, 2)) # split into 2 along columns (horizontal split)
print(np.vsplit(s, 2)) # split into 2 along rows (vertical split)
```
